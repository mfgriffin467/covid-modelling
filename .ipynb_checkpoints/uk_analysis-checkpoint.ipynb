```python
import numpy as np
import pandas as pd
import folium
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read and process data

```python
xls = pd.ExcelFile('https://fingertips.phe.org.uk/documents/Historic%20COVID-19%20Dashboard%20Data.xlsx')
df_raw = pd.read_excel(xls, 'UTLAs', skiprows=7)
latest_date = df_raw.columns[-1]
df_country = pd.read_excel(xls, 'Countries', skiprows=7).drop(['Area Code'], axis=1)
pop = pd.read_csv("data_feeds/population_stats.csv")
df_map = pd.read_csv("data_feeds/Ward_to_Local_Authority_District_to_County_to_Region_to_Country_December_2018_Lookup_in_United_Kingdom_.csv")
df_map2 = pd.read_csv("data_feeds/Historic_COVID19_Dashboard_Data.csv")
```

```python
geo = 'https://opendata.arcgis.com/datasets/658297aefddf49e49dcd5fbfc647769e_4.geojson'

test2 = pd.read_json(geo)
output = []
for i in range(0,len(test2)):
    new = test2.iloc[i].features['properties']['ctyua17nm']
    output.append(new)
```

```python
df_total = df_raw.merge(df_map2[['Area Name', 'NHS region']], on='Area Name')
df_total['area_map'] = df_total['Area Name'].apply(lambda x: 'Cornwall' if x == 'Cornwall and Isles of Scilly' \
                                                     else 'Hackney' if x == 'Hackney and City of London' \
                                                     else 'Bournemouth' if x == 'Bournemouth, Christchurch and Poole' \
                                                     else x)
df_total = df_total.merge(pop[['Name','population_estimate']], how = 'left', left_on='area_map', right_on='Name').drop(['Name'],axis=1)
df_total['case_density'] = df_total[latest_date]/df_total['population_estimate']*100000
```

```python
# Create average entries for Scotland, Wales and NI

df_visual = df_total[['area_map', 'case_density', latest_date]]
df_other = df_map[df_map.CTRY18NM != 'England'][['CTRY18NM','LAD18NM']].groupby(['CTRY18NM','LAD18NM']).count().reset_index()
df_other.columns = ['country','area_map']
df_comb = df_other.merge(df_country, how='left', left_on='country', right_on='Area Name')
df_nations = df_other.groupby(['country']).count().reset_index()
df_nations.columns = ['country', 'nations_area_count']
df_comb = df_comb.merge(df_nations, how='left', left_on='country', right_on='country')
df_comb[latest_date] = df_comb[latest_date]/df_comb['nations_area_count']
df_comb = df_comb.merge(pop[['Name','population_estimate']], how = 'left', left_on='area_map', right_on='Name').drop(['Name'],axis=1)
ctry_summary = df_comb[['country','population_estimate', latest_date]].groupby('country').sum()
ctry_summary['case_density'] = ctry_summary[latest_date]/ctry_summary['population_estimate']*100000
ctry_summary = ctry_summary.reset_index()
df_comb = df_comb.merge(ctry_summary[['country','case_density']], how = 'left', on='country')
df_visual = pd.concat([df_visual, df_comb], axis=0, sort=True)
```

### Time series

```python
# Region splits

region_summary = df_total.drop(['population_estimate','case_density'],axis=1).groupby('NHS region').sum().T
region_summary.index = pd.to_datetime(region_summary.index, format='%Y-%m-%d %X' )
region_summary.index = region_summary.index.date
plt.figure(figsize=(16,8))
ax = sns.lineplot(data=region_summary, dashes=False)
ax.set_title("Confirmed cases by county")
plt.show()
```

```python
# Local authority splits

top_areas = df_total.sort_values(by=latest_date, ascending=False)['Area Name'].head(10)

region_summary = df_total.drop(['population_estimate','case_density'],axis=1).groupby('Area Name').sum().T
region_summary.index = pd.to_datetime(region_summary.index, format='%d/%m/%Y')
plt.figure(figsize=(16,8))
ax = sns.lineplot(data=region_summary[top_areas], dashes=False)
ax.set_title("Confirmed cases for top 10 local authorities")
plt.show()
```

### Map visuals

```python
# Raw counts

my_map2 = folium.Map(location=[54,0],zoom_start = 6) # for UK

chloropleth = folium.Choropleth(
    geo_data=geo,
    data=df_visual,
    columns = ['area_map',latest_date],
    key_on='feature.properties.ctyua17nm',
    fill_color='PuRd',#OrRd
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Cases per area',
    highlight=True
).add_to(my_map2)

chloropleth.geojson.add_child(folium.features.GeoJsonTooltip(['ctyua17nm'], labels=False))

#my_map2.save(path + 'dashboard_totals.html')
my_map2
```

```python
# Density

my_map2 = folium.Map(location=[54,0],zoom_start = 6) # for UK

chloropleth = folium.Choropleth(
    geo_data=geo,
    data=df_visual,
    columns = ['area_map','case_density'],
    key_on='feature.properties.ctyua17nm',
    fill_color='PuRd',#OrRd
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Cases per area',
    highlight=True
).add_to(my_map2)

chloropleth.geojson.add_child(folium.features.GeoJsonTooltip(['ctyua17nm'], labels=False))

#my_map2.save(path + 'dashboard_totals.html')
my_map2
```
